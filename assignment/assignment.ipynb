```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/UVA/24F/ML/college_completion.csv',low_memory=False)
```

```python
df.head()
```

```python
print(df.shape, '\n') # List the dimensions of df
print(df.dtypes, '\n') # The types of the variables; `object` is a bad sign
print(df.columns, '\n') # Column names
```

```python
control = df["control"]
level = df["level"]

print(pd.crosstab(control, level),'\n')
```

```python
# Statistical description
print("Statistical Description:")
print(df.describe())

# Plot Histogram
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
plt.hist(df['grad_100_value'], bins=30, edgecolor='black')
plt.title("Histogram")

# Plot Kernel Density Plot
plt.subplot(2, 2, 2)
sns.kdeplot(df['grad_100_value'], shade=True)
plt.title("Kernel Density Plot")

# Plot Boxplot
plt.subplot(2, 2, 3)
sns.boxplot(df['grad_100_value'])
plt.title("Boxplot")

plt.tight_layout()
plt.show()
```

```python
# Grouped Kernel Density Plot by 'control' and 'level'
plt.figure(figsize=(10, 6))

# KDE plot by 'control'
plt.subplot(2, 1, 1)
sns.kdeplot(data=df, x='grad_100_value', hue='control', shade=True)
plt.title("Kernel Density Plot by Control")

# KDE plot by 'level'
plt.subplot(2, 1, 2)
sns.kdeplot(data=df, x='grad_100_value', hue='level', shade=True)
plt.title("Kernel Density Plot by Level")

plt.tight_layout()
plt.show()

# Grouped Statistical Descriptions
grouped_stats = df.groupby(['control', 'level'])['grad_100_value'].describe()
print("Grouped Statistical Descriptions:")
print(grouped_stats)
```

```python
df['levelXcontrol']=df['level']+', '+df['control']
df.head()
```

```python
# Grouped Kernel Density Plot by 'control' and 'level'
plt.figure(figsize=(10, 6))

# KDE plot by 'control'
plt.subplot(2, 1, 1)
sns.kdeplot(data=df, x='aid_value', hue='control', shade=True)
plt.title("Kernel Density Plot by Control")

# KDE plot by 'level'
plt.subplot(2, 1, 2)
sns.kdeplot(data=df, x='aid_value', hue='level', shade=True)
plt.title("Kernel Density Plot by Level")

plt.tight_layout()
plt.show()

# Grouped Statistical Descriptions
grouped_stats = df.groupby(['control', 'level'])['aid_value'].describe()
print("Grouped Statistical Descriptions:")
print(grouped_stats)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='aid_value', y='grad_100_value')
plt.title("Scatterplot of Graduation Rate by Aid Value")
plt.xlabel("Aid Value")
plt.ylabel("Graduation Rate (grad_100_value)")
plt.show()
```

```python
# Scatterplot grouped by 'level'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='aid_value', y='grad_100_value', hue='level')
plt.title("Scatterplot of Graduation Rate by Aid Value, Grouped by Level")
plt.xlabel("Aid Value")
plt.ylabel("Graduation Rate (grad_100_value)")
plt.legend(title='Level')
plt.show()

# Scatterplot grouped by 'control'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='aid_value', y='grad_100_value', hue='control')
plt.title("Scatterplot of Graduation Rate by Aid Value, Grouped by Control")
plt.xlabel("Aid Value")
plt.ylabel("Graduation Rate (grad_100_value)")
plt.legend(title='Control')
plt.show()
```

Based on the second scatterplot (Grad Rate by Aid Value, Grouped by Control) it seems that for Private not-for-profit schools if more aid is given the graduation rate is higher (there is a general upward trend in the data). By looking at the first scatterplot (Grad Rate by Aid Value, Grouped by Level) for 4-year colleges, more aid given corresponds to a higher graduation rate.

```python
# -- 3 --
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/UVA/24F/ML/sfpuf2021_1_fall.csv',low_memory=False)
df.head()
```

```python
# -- Dual Eligibility --
var1 = 'ADM_OP_MDCD'
print("Dual Eligibility:", df[var1].describe(),'\n')


# -- Low Income Subsidy --
var2 = 'ADM_LIS_FLAG_YR'
print("Low Income Subsify:", df[var2].describe(),'\n')
```

```python
# -- Dual Eligibility --
df['ADM_OP_MDCD_nan'] = df['ADM_OP_MDCD'].equals('.')
missing = 0;
for i in range(len(df['ADM_OP_MDCD_nan'])):
  if df['ADM_OP_MDCD_nan'][i] == True:
    missing += 1
print('Total Missings: \n', missing,'\n')


# -- Low Income Subsidy --
df['ADM_LIS_FLAG_YR_nan'] = df['ADM_LIS_FLAG_YR'].equals('.')
missing = 0;
for i in range(len(df['ADM_LIS_FLAG_YR_nan'])):
  if df['ADM_LIS_FLAG_YR_nan'][i] == True:
    missing += 1
print('Total Missings: \n', missing,'\n')
```

Neither variable has any missing values because there are only numerical codes and no '.'; this is confirmed when looking in the data dictionary.

```python
import pandas as pd
import matplotlib.pyplot as plt

# dataframe for ADM_OP_MDCD; Counts_MDCD comes from data dictionary
mdcd_data = {
    'ADM_OP_MDCD': ['Nondual', 'Full dual', 'Partial dual', 'QMB dual'],
    'Counts_MDCD': [10111, 1835, 400, 437]
}
df_mdcd = pd.DataFrame(mdcd_data)

## dataframe for ADM_LIS_FLAG_YR; Counts_LIS comes from data dictionary
lis_data = {
    'ADM_LIS_FLAG_YR': ['No premium subsidy', 'Part year premium subsidy', 'Full year premium subsidy'],
    'Counts_LIS': [9752, 192, 2839]
}
df_lis = pd.DataFrame(lis_data)

# ADM_OP_MDCD bar graph
plt.figure(figsize=(10, 5))
plt.bar(df_mdcd['ADM_OP_MDCD'], df_mdcd['Counts_MDCD'], color='skyblue')
plt.title('Medicare-Medicaid Dual Eligibility Counts')
plt.xlabel('Medicare-Medicaid Dual Eligibility')
plt.ylabel('Counts')
plt.show()

# ADM_LIS_FLAG_YR bar graph
plt.figure(figsize=(10, 5))
plt.bar(df_lis['ADM_LIS_FLAG_YR'], df_lis['Counts_LIS'], color='lightgreen')
plt.title('Low Income Subsidy Indicator for the Year')
plt.xlabel('Low Income Subsidy')
plt.ylabel('Counts')
plt.show()
```

```python
# histogram for ADM_OP_MDCD
plt.figure(figsize=(10, 5))
plt.hist(df_mdcd['Counts_MDCD'], bins=4, color='skyblue', edgecolor='black')
plt.title('Histogram of Medicare-Medicaid Dual Eligibility Counts')
plt.xlabel('Counts')
plt.ylabel('Frequency')
plt.show()

# histogram for ADM_LIS_FLAG_YR
plt.figure(figsize=(10, 5))
plt.hist(df_lis['Counts_LIS'], bins=3, color='lightgreen', edgecolor='black')
plt.title('Histogram of Low Income Subsidy Indicator Counts')
plt.xlabel('Counts')
plt.ylabel('Frequency')
plt.show()
```
